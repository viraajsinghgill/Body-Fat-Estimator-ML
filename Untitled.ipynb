```python
def plotdistplot(col):
    
    plt.figure(figsize=(12,4))
    sn.distplot(df['BodyFat'],color='m',hist=False,label='BodyFat')
    sn.distplot(df[col],color='r',hist=False,label=col)
    plt.legend()
    plt.show()
    
col=list(df.columns)
for i in col:
    print(f'Distribution plots  for {i} feature shown below')
    plotdistplot(i)
    print("-"*100)
```

```python
import scipy.stats as stats
```

```python
def drawplots(df,col):
    
    plt.figure(figsize=(15,7))
    plt.subplot(1,3,1)
    plt.hist(df[col],color='m')
    
    plt.subplot(1,3,2)
    stats.probplot(df[col],dist='norm',plot=plt)
    
    plt.subplot(1,3,3)
    sn.boxplot(df[col],color='m')
    
    plt.show()

cols=list(df.columns)
for i in range(len(cols)):
    print(f'Distribution plots for the feature {cols[i]} are shown below')
    drawplots(df,cols[i])
    print("-"*100)
```

```python
# checking for ouliers
upperlimit=[]
lowerlimit=[]
for i in df.columns:
    upperlimit.append(df[i].mean()+df[i].std()*4)
    lowerlimit.append(df[i].mean()-df[i].std()*4)
```

```python
cols=list(df.columns)
j=0
for i in range(len(cols)):
    temp=df.loc[(df[cols[i]]>upperlimit[j])&(df[cols[i]]<lowerlimit[j])]
    j+=1
```

```python
def correlation(df,threshold):
    colcor=set()
    
    cormat=df.corr()
    
    for i in range(len(cormat)):
        for j in range(i):
            if abs(cormat.iloc[i][j])>threshold:
                colname=cormat.columns[i]
                colcor.add(colname)
                
    return colcor
ans=correlation(train,threshold=0.85)

ans
    
```

```python
totrain=train[cols2]
totrain.head()
```

```python
from sklearn.model_selection import train_test_split,RandomizedSearchCV
from sklearn import metrics
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression,Lasso,Ridge
```

```python
X_train,X_test,y_train,y_test=train_test_split(totrain,test,test_size=0.2)
```

```python
reg=DecisionTreeRegressor()
reg.fit(X_train,y_train)
plt.figure(figsize=(15,7))
tree.plot_tree(reg,filled=True)
```

```python
path=reg.cost_complexity_pruning_path(X_train,y_train)
ccp_alpha=path.ccp_alphas
```

```python
alphalist=[]
for i in range(len(ccp_alpha)):
    reg=DecisionTreeRegressor(ccp_alpha=ccp_alpha[i])
    reg.fit(X_train,y_train)
    alphalist.append(reg)
```

```python
trainscore=[alphalist[i].score(X_train,y_train) for i in range(len(alphalist))]
            
testscore=[alphalist[i].score(X_test,y_test) for i in range(len(alphalist))]
            
plt.xlabel('alpha')
plt.ylabel('accuracy')
plt.plot(ccp_alpha,trainscore,marker='o',label='training',color='m',drawstyle='steps-post')
plt.plot(ccp_alpha,testscore,marker='+',label='testing',color='r',drawstyle='steps-post')
plt.legend()
plt.show()
```

```python
clf=DecisionTreeRegressor(ccp_alpha=1)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
print(f'Decision tree normal approach:{metrics.r2_score(y_test,y_pred)}')


rf=RandomForestRegressor(n_estimators=1000,ccp_alpha=1)
rf.fit(X_train,y_train)
y_pred_rf=rf.predict(X_test)
print(f'random foreste normal approach:{metrics.r2_score(y_test,y_pred_rf)}')
```
